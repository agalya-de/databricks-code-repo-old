```python

%python

df1=spark.read.csv("/Volumes/workspace/default/usage_metrics/mobile_os_usage.csv",header=True)
print(df1.is_cached)
df2=df1.repartition(4)
display(df2)
```

```python
%python
#Task7: Write the above dataframe df1 data into a databricks table called 'default.mobile_os_usage' using pyspark magic command

# Clean column names: replace spaces with underscores
df1_clean = (
    df1
    .withColumnRenamed("Mobile Operating System", "mobile_operating_system")
    .withColumnRenamed("Percent of Usage", "percent_of_usage")
)

# Save as Delta table
df1_clean.write.mode("overwrite").saveAsTable("default.mobile_os_usage")

print("Table 'default.mobile_os_usage' created successfully.")
```

```python
%python
#Task8: Write sql query to display the data loaded into the table 'default.mobile_os_usage' using the pyspark magic command 

display(spark.sql("select * from default.mobile_os_usage limit 10"))
```
